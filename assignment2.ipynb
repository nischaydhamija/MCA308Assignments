```python
import pandas as pd

# Create DataFrame
employees = pd.DataFrame({
"Employee_ID":[101,102,103,104,105],
"Name":["Alice","Bob","Charlie","Diana","Edward"],
"Department":["HR","IT","IT","Marketing","Sales"],
"Age":[29,34,41,28,38],
"Salary":[50000,70000,65000,55000,60000],
"Years_of_Experience":[4,8,10,3,12],
"Joining_Date":["2020-03-15","2017-07-19","2013-06-01","2021-02-10","2010-11-25"],
"Gender":["Female","Male","Male","Female","Male"],
"Bonus":[5000,7000,6000,4500,5000],
"Rating":[4.5,4.0,3.8,4.7,3.5]
})

employees.shape

employees.info()

employees.describe()

employees.head()
employees.tail(3)

employees["Salary"].mean()
employees["Bonus"].sum()
employees["Age"].min()
employees["Rating"].max()

employees.sort_values(by="Salary", ascending=False)

employees["Category"] = employees["Rating"].apply(
    lambda r: "Excellent" if r>=4.5 else ("Good" if r>=4 else "Average"))

employees.isnull().sum()

employees = employees.rename(columns={"Employee_ID":"ID"})

employees[employees["Years_of_Experience"]>5]
employees[employees["Department"]=="IT"]


employees["Tax"] = employees["Salary"] * 0.10


employees.to_csv("employees_updated.csv", index=False)
```
