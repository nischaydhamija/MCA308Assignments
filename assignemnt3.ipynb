```python
import pandas as pd
df = pd.read_csv("AdventureWorks_sample.csv")   # change name if your file differs
df.head()
```

```python
selected_cols = [
    "Age","Gender","Income","MaritalStatus","NumberChildrenAtHome",
    "Education","Occupation","HomeOwnerFlag","NumberCarsOwned",
    "CommuteDistance","BikeBuyer"
]

df_selected = df[selected_cols]
df_selected.head()
```

```python
df_selected.dtypes
```

```python
df_selected.groupby("BikeBuyer")["Income"].mean()
```

```python
df_selected.groupby("CommuteDistance")["BikeBuyer"].mean()*100
```

```python
import matplotlib.pyplot as plt

df_selected.boxplot(column="Age", by="BikeBuyer")
plt.title("Age vs Bike Buyer")
plt.suptitle("")
plt.show()
```

```python
df_selected.groupby("MaritalStatus")["BikeBuyer"].mean()*100
```
