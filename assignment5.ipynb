```python
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, hamming
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import jaccard_score

customer_A = np.array([4, 5, 2, 3, 4])
customer_B = np.array([5, 3, 2, 4, 5])

customer_A_binary = np.array([1, 0, 1, 1, 0, 1])
customer_B_binary = np.array([1, 1, 1, 0, 0, 1])
```

```python
euclid_dist = euclidean(customer_A, customer_B)
print("Euclidean Distance:", euclid_dist)
```

```python
manhattan_dist = cityblock(customer_A, customer_B)
print("Manhattan Distance:", manhattan_dist)
```

```python
cos_sim = cosine_similarity([customer_A], [customer_B])[0][0]
print("Cosine Similarity:", cos_sim)
```

```python
hamm_dist = hamming(customer_A_binary, customer_B_binary)
print("Hamming Distance:", hamm_dist)
```

```python
jaccard_sim = jaccard_score(customer_A_binary, customer_B_binary)
print("Jaccard Similarity:", jaccard_sim)
```
